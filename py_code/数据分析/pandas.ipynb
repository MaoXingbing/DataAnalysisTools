Pandas主要有两个数据结构Series和DataFrame

```python
#Series对象是一个一维数组的数据结构，可以存储整数，浮点数，结构由一组索引和与之对应的数据组成，其中索引默认为整数，从0开始
import pandas as pd
import numpy as np
a1=pd.Series([1,2,3,4])
print("使用列表创建的Series:\n",a1)
a2=np.array([2,2,2,2])
a3=pd.Series(a2,['a','b','c','d'])
print("有索引规范的Series：",a3)
a4=pd.Series({'a':1,'b':2})
print('使用字典创建Series对象:\n',a4)
```

```python
#DataFrame对象一个类似二维数组的数据结构，每列是一个Series对象，且可以有不同的数据类型，DataFrame由索引与数据组成，不仅有行索引还是有列索引
index=['wang','tidy','pite','lee']  #定义行标签
columns=['math','PE','art']  #定义列标签
dict={'math':[95,1,34,67],'PE':[2,23,25,45],'art':[54,98,65,36]}
df1=pd.DataFrame(dict,index)
print(df1)
```
